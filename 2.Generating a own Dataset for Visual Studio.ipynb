```python
# Python Code to Generate the own dataset


import pandas as pd
import random
from datetime import datetime, timedelta

# Setup
juice_types = ["Orange", "Apple", "Mango", "Pineapple", "Watermelon", "Grape", "Pomegranate", "Mixed Fruit"]
payment_methods = ["Cash", "Card", "UPI", "Wallet"]
times_of_day = ["Morning", "Afternoon", "Evening"]
customer_types = ["New", "Regular"]
genders = ["Male", "Female"]

# Generate 500+ rows
start_date = datetime(2025, 7, 1)
data = []

for _ in range(545):
    date = start_date + timedelta(days=random.randint(0, 30))
    juice = random.choice(juice_types)
    quantity = random.randint(1, 5)
    unit_price = random.randint(40, 70)
    total_price = quantity * unit_price
    age = random.randint(10, 65)
    gender = random.choice(genders)
    payment = random.choice(payment_methods)
    time = random.choice(times_of_day)
    cust_type = random.choice(customer_types)
    

    data.append({
        "Date": date.strftime('%Y-%m-%d'),
        "Juice Type": juice,
        "Quantity": quantity,
        "Unit Price": unit_price,
        "Total Price": total_price,
        "Customer Age": age,
        "Gender": gender,
        "Payment Method": payment,
        "Time of Day": time,
        "Customer Type": cust_type,
        
    })

# Convert to DataFrame and save
df = pd.DataFrame(data)
df.to_csv("Juice_Sales_July2025_VisualStudio.csv", index=False)

print("csv file created successfully!")
```
