# jointplot

# Jointplot Is the main function In Python visualisation library its used to find the relations between the 2 variable


A joint plot is a versatile visualization tool from the Seaborn library in Python. It combines two types of plots into a single figure to help you explore the relationship between two variablesâ€”for example, height vs. weight or age vs. income.

- The main plot (usually a scatter plot) shows how the two variables relate to each other.

- The side plots display the individual distributions of each variableâ€”typically as histograms or KDE curves.

- Scatter Plot: Displays data points to show correlation
- Hex Plot: Uses hexagonal bins to show density
- Contour Plot: Shows smooth density contours for continuous data
- x & y: variables to compare


```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x = 'ssc_p',y = 'salary',data = dataset,kind='kde')
plt.show()
```

#         Factorplot

- factorplot was a high-level plotting function in Seaborn used to visualize categorical data. It allowed you to quickly generate plots like bar charts, box plots, and point plots based on categorical groupings.
- However, in newer versions of Seaborn, factorplot has been deprecated and replaced by catplot, which offers the same functionality with more flexibility


# Factor
- In statistics, a factor is a categorical variable. So factorplot was designed to show how a numeric variable changes across different categories (factors)



```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('exercise')
sb.factorplot(x = "time", y = "pulse", hue = "kind", kind = 'violin', col = "diet", data = df);
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
sb.regplot(x = "total_bill", y = "tip", data = df)
sb.lmplot(x = "total_bill", y = "tip", data = df)
plt.show()
```
